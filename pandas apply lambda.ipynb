### Challenge 4
Now the real stuff

We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers.

```python
def is_prime(n):
    """
    Function to check if a number is prime.
    """
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
```

```python
# combruebo que funcione
is_prime(11)
```

```python
is_prime(15)
```

```python
# defino la primera funcion que sumará los ascii de los títulos en una columna
df['sum_ascii'] = df['Title'].apply(lambda x: sum(ord(c) for c in x))
```

```python
# defino la columna que dividira la suma de ascii entre los votos de películas
df['sum_div'] = df.apply(lambda x : revenue_minute(x['sum_ascii'], x['Votes']), axis=1)
```

```python
# aplico la función con los resultados en otra columna
df['is_prime'] = df['sum_div'].apply(lambda x: is_prime(int(x)))
```

### Challenge 6. Freaky bonus
We want to know which movies might have hidden paterns in their description. A way to know that is finding those movies which the sum of all numeric values of the string description hash (SHA256) are between their revenue and their number of votes.

```python
# no lo entiendo ni con chatGPT
```

```python
import hashlib
import pandas as pd


# Define a function to extract all numeric values from a string
def extract_numeric_values(string):
    numeric_values = []
    for char in string:
        if char.isdigit():
            numeric_values.append(int(char))
    return numeric_values

# Calculate the SHA256 hash of the Description column for each movie
hashes = []
for desc in df["Description"]:
    hash_object = hashlib.sha256(desc.encode())
    hash_value = hash_object.hexdigest()
    hashes.append(hash_value)

# Extract all numeric values from the hash for each movie and sum them up
numeric_sums = []
for hash_value in hashes:
    numeric_values = extract_numeric_values(hash_value)
    numeric_sum = sum(numeric_values)
    numeric_sums.append(numeric_sum)

# Check if the sum of numeric values falls between the Revenue and Votes columns for each movie
for i in range(len(df)):
    if df.iloc[i]["Revenue (Millions)"] <= numeric_sums[i] <= df.iloc[i]["Votes"]:
        print(df.iloc[i]["Title"])
df
```
